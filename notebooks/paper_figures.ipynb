```python
def plot_bar_and_line(bar_data, line_data):
    fig, ax1 = plt.subplots()

    # Create the bar plot
    bar_color = '#498467'
    ax1.bar(range(len(bar_data)), bar_data, color=bar_color)
    ax1.set_ylabel('ASAT R2 Score', color=bar_color)
    ax1.tick_params(axis='y', labelcolor=bar_color)
    ax1.set_ylim(0.92, 0.944)

    # Create the line plot
    line_color= 'orange'
    ax2 = ax1.twinx()
    ax2.plot(range(len(line_data)), line_data, color=line_color, linewidth=5)
    ax2.set_ylabel('Training time', color=line_color)
    ax2.tick_params(axis='y', labelcolor=line_color)

    # Set x-axis ticks and labels
    ax1.set_xticks(range(len(bar_data)))
    ax1.set_xticklabels(range(len(bar_data)))

    # Set the title and labels
    # plt.title('Bar Plot and Line Plot')
    plt.xlabel('Decimation')

    # Display the figure
    plt.show()
```
